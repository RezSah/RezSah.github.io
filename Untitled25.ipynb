Here we want to investigate about movies released in the 90s.

```python
# Filtered based on movies released in the 90s.
movies_1990s = df.loc[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]

```

```python
# Counting the number of movies released in the 90s.
print(movies_1990s['show_id'].count())
```

Let's count the number of movies shorter than 90 minutes and call it short movies.

```python
short_movies = movies_1990s[movies_1990s['duration'] < 90]
```

```python
print(short_movies['show_id'].count())
```

```python
sns.histplot(data = movies_1990s , x = 'duration' ,bins = 10, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations in 90s')
plt.xlabel('Number of Movies')
plt.ylabel('Duration (minutes)')
plt.show()
```
