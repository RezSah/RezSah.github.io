Here we want to investigate about movies released in the 90s.

```python
# Filtered based on movies released in the 90s.
movies_1990s = df.loc[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]

```

```python
# Counting the number of movies released in the 90s.
print(movies_1990s['show_id'].count())
```

Let's count the number of movies shorter than 90 minutes and call it short movies.

```python
short_movies = movies_1990s[movies_1990s['duration'] < 90]
```

```python
print(short_movies['show_id'].count())
```

```python
sns.histplot(data = movies_1990s , x = 'duration' ,bins = 10, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations in 90s')
plt.xlabel('Number of Movies')
plt.ylabel('Duration (minutes)')
plt.show()
```

Now we want to investigate how many short movies were released in the 90s that were.


```python
count_of_movies_per_country = movies_1990s.groupby('country')['show_id'].count().reset_index()
```

```python
count_of_movies_per_country.columns = ['country', 'count_of_movies']
```

```python
print(count_of_movies_per_country)
```

```python
categories = movies_1990s['country']
values = count_of_movies_per_country

# Create a bar chart
sns.barplot(x= 'country', y= 'count_of_movies', data= count_of_movies_per_country, palette='pastel')

plt.title('Number of Movies in the 90s Based on Country')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=90)
plt.show()
```

```python
count_type_of_film = movies_1990s.groupby('type')['show_id'].count().reset_index()
```

```python
count_type_of_film
```

```python
count_type_of_film.columns = ['type','count_type_of_film']
```

```python
type_counts = movies_1990s.groupby('type')['show_id'].count()
categories = movies_1990s.groupby('type')['show_id'].count()
values = movies_1990s.groupby('type')['show_id'].count().values

plt.pie(values, labels=categories)  # No autopct for percentages

# Display the raw sizes as labels
plt.title('Count of Movie Types in the 90s')
plt.axis('equal')
plt.show()
```

```python
count_of_movies_per_director = movies_1990s.groupby('director')['title'].count()
```

```python
count_of_movies_per_director.sort_values(ascending = False)
```

```python
dropping_unknowns = movies_1990s[movies_1990s['director'] != 'Unknown']
```

```python
directors_most_movie = dropping_unknowns.groupby('director')['title'].count().sort_values(ascending = False)
```

```python
directors_most_movie.head(5)
```

```python
movie_duration = movies_1990s.pivot_table(values='duration', index='title')
```

```python
movie_duration.sort_values(by='duration',ascending=False).head(10)
```

```python
categories = movies_1990s.groupby('genre')['show_id'].count()
top_5_genres = categories.sort_values(ascending=False).head(5)
values = top_5_genres.values
labels = top_5_genres.index
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, startangle=50)
plt.axis('equal')
plt.title('The Most Popular Genres of the 90s')
plt.show()
```
